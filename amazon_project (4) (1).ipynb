```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data cleaning below we has raw data

```python
df=pd.read_csv('/content/Amazon Sale Report.csv')
df
```

# 2nd question Droping the Columns to drop() Unnamed: 22 - undeterminable data

```python

df = df.drop('Unnamed: 22', axis=1)
```

```python
df
```

#1 QUESTION SETTING DATE AND TIME FROM APRIL 1ST OF 2022 TO 29TH JUNE 2022

```python
df['Date']=pd.to_datetime(df["Date"])
s='2022-04-01'
e='2022-06-29'
df=df.query('Date>=@s and Date<=@e')
df
```

# 3RD QUESTION FILTERING THE COLUMNS
   FULLFILLMENT = EASYSHIP

```python

df = df.drop('fulfilled-by', axis=1)
df
```

#4TH QUESTION FILTERING THE COLUMN
SHIP COUNTRY=INDIA

```python

#df = df.drop('ship-country', axis=1)
df
```

#5TH QUESTION FLITERING COLUMN
 CURRENY= INR

```python

#df = df.drop('Sales Channel', axis=1)
#df
```

#6TH QUESTION REMOVING COLUMN DUPLICATION WITH COLUMNS COMBINATION

ORDER ID AND PRODUCT

```python
df = df.drop_duplicates(subset=['Order ID', 'ASIN'])
df
```

# 7TH QUESTION FILLING NAN VALUES WITH UNKNOWN IN ORDER STATUS COLUMN

```python
df['Courier Status'] = df['Courier Status'].fillna('Unknown')
df
```

# 8TH QUESTION FILLING NAN VALUES WITH nopromotion IN  promotionids COLUMN

```python
df['promotion-ids'] = df['promotion-ids'].fillna('No promotion')
df
```

#9TH QUESTION FILLING AMOUNT WITH ZERO VALUE WHERE THERE IS NULL

```python
df['Amount'] = df['Amount'].fillna(value=0)
df
```

#10TH QUESTION CHANGING THE NAMES OF THE COLUMNS


```python
df = df.rename(columns={'B2B': 'Customer_type', 'Amount': 'Order_amount_($)'})
df
```

```python
replacement_dict = {'False':'customer', 'True': 'business'}
df['Customer_type'] = df['Customer_type'].replace(replacement_dict)
df
```

#11TH QUESTION  CONVERTING THE ORDER AMOUNT FROM INR TO DOLLAR

```python
exchange_rate=0.0120988
df['Order_amount_($)']= (df['Order_amount_($)'].apply(lambda x:x * exchange_rate)).round(2)
df
```

#12TH QUESTION TO REMOVE THE MARCH DATE SINCE THERE IS ONLY ONE DAY IN THE DATA

```python
df = df.loc[df['Date'].dt.month != 3]
df
```

#13th creating new column as month to see the month in word format

```python
df['month'] = df['Date'].dt.strftime('%B')
df
```

```python
df['Customer_type'].replace({False: 'business', True:'customer'}, inplace=True)
df
```

```python
SIZE = df['Size'].unique
SIZE
```

#14th Creating the size column in order wise by using catergorical function

```python
sizeorder=['Free','XS','S','M','L','XL','XXL','3XL','4XL','5XL','6XL']
df['Size']=pd.Categorical(df['Size'],categories=sizeorder,ordered=True)
df
```

# Analysis of state wise revenue

```python
revenue_statewise = df.groupby('ship-state')['Order_amount_($)'].sum()
plt.figure(figsize=(10, 6))
revenue_statewise.plot(kind='bar', color='skyblue')
plt.title('State-wise Revenue Distribution')
plt.xlabel('State')
plt.ylabel('Total Revenue')
plt.show()
```

# top 50% of states revenue wise


```python
# top 50% of states revenue wise
revenue_statewise = df.groupby('ship-state')['Order_amount_($)'].sum()

cumulative_percentage = revenue_statewise.sort_values(ascending=False).cumsum() / revenue_statewise.sum()

top50_states = cumulative_percentage[cumulative_percentage <= 0.5].index

df_top50_states = df[df['ship-state'].isin(top50_states)]

plt.figure(figsize=(6, 6))
plt.bar(df_top50_states['ship-state'], df_top50_states['Order_amount_($)'], color='skyblue')
plt.title('Top 50% State-wise Revenue Distribution')
plt.xlabel('State')
plt.ylabel('Total Revenue')

plt.show()
```

# bottom 50% of states revenue wise


```python
cumulative_percentage = revenue_statewise.sort_values(ascending=True).cumsum() / revenue_statewise.sum()

bottom50_states = cumulative_percentage[cumulative_percentage <= 0.5].index


df_bottom50_states = df[df['ship-state'].isin(bottom50_states)]

plt.figure(figsize=(10, 6))
plt.bar(df_bottom50_states['ship-state'], df_bottom50_states['Order_amount_($)'], color='skyblue')
plt.title('Bottom 50% State-wise Revenue Distribution')
plt.xlabel('State')
plt.ylabel('Total Revenue')
plt.xticks(rotation='vertical')
plt.show()
```
